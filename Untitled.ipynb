```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
#headers = headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36'}
url="https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&otracker=categorytree&page=1"
response=urlopen(url)
response_data=response.read()
```

```python
data=soup(response_data)
data
```

```python
div_data=data.findAll('div',{'class':'_2kHMtA'})
print(len(div_data))
```

# Single Page

```python
product=[]
star=[]
rating=[]
review=[]
pro=[]
ram=[]
rom=[]
display=[]
current_price=[]
mrp=[]
img_link=[]



for i in div_data:
    tit=i.findAll('div',{'class':'_4rR01T'})
    product.append(tit[0].text.split('-')[0].strip())
    try:
        st=i.find('div',{'class':'_3LWZlK'})
        star.append(st.text)
    except:
        star.append('0')
    try:
        rating_span=i.findAll('span',{'class':'_2_R_DZ'})
        rating.append(rating_span[0].text.split('&')[0].replace("Ratings",'').strip())
    except:
        rating.append(0)
    try:
        review_span=i.findAll('span',{'class':'_2_R_DZ'})
        review.append((review_span[0].text.split('&')[1].replace("Reviews",'').strip()))
    except:
        review.append(0)
    info=i.findAll('li',{"class":'rgWa7D'})
    pro.append(info[0].text)
    info=i.findAll('li',{"class":'rgWa7D'})
    ram.append(info[1].text)
    info=i.findAll('li',{"class":'rgWa7D'})
    rom.append(info[3].text)
    info=i.findAll('li',{"class":'rgWa7D'})
    display.append(info[4].text)
    try:
        cp=i.find('div',{'class':'_30jeq3 _1_WHN1'})
        current_price.append(cp.text)
    except:
        current_price.append(0)
    try:
        mp=i.find('div',{'class':'_3I9_wc _27UcVY'})
        mrp.append(mp.text)
    except:
        mrp.append(0)
    
    image=i.findAll('div',{'class':'CXW8mj'})
    img_link.append(image[0].img.get('src'))
```

# Scrape multiple pages

```python
final=pd.DataFrame()
for page in np.arange(1,42):
    url="https://www.flipkart.com/laptops/pr?sid=6bo%2Cb5g&otracker=categorytree&page={}".format(page)
    response=urlopen(url)
    response_data=response.read()
    data=soup(response_data)
    div_data=data.findAll('div',{'class':'_2kHMtA'})
    product=[]
    star=[]
    rating=[]
    review=[]
    pro=[]
    ram=[]
    rom=[]
    display=[]
    current_price=[]
    mrp=[]
    img_link=[]



    for i in div_data:
        tit=i.findAll('div',{'class':'_4rR01T'})
        product.append(tit[0].text.split('-')[0].strip())
        try:
            st=i.find('div',{'class':'_3LWZlK'})
            star.append(st.text)
        except:
            star.append('0')
        try:
            rating_span=i.findAll('span',{'class':'_2_R_DZ'})
            rating.append(rating_span[0].text.split('&')[0].replace("Ratings",'').strip())
        except:
            rating.append(0)
        try:
            review_span=i.findAll('span',{'class':'_2_R_DZ'})
            review.append((review_span[0].text.split('&')[1].replace("Reviews",'').strip()))
        except:
            review.append(0)
        info=i.findAll('li',{"class":'rgWa7D'})
        pro.append(info[0].text)
        info=i.findAll('li',{"class":'rgWa7D'})
        ram.append(info[1].text)
        info=i.findAll('li',{"class":'rgWa7D'})
        rom.append(info[3].text)
        info=i.findAll('li',{"class":'rgWa7D'})
        display.append(info[4].text)
        try:
            cp=i.find('div',{'class':'_30jeq3 _1_WHN1'})
            current_price.append(cp.text)
        except:
            current_price.append(0)
        try:
            mp=i.find('div',{'class':'_3I9_wc _27UcVY'})
            mrp.append(mp.text)
        except:
            mrp.append(0)

        image=i.findAll('div',{'class':'CXW8mj'})
        img_link.append(image[0].img.get('src'))
    dict={'Product Title':product,"Star":star,"Rating":rating,"Review":review,"Processor":pro,"RAM":ram,"ROM":rom,"Screen Size":display,"Current Price":current_price,"MRP":mrp,"Image Link":img_link}
    df=pd.DataFrame(dict)
    final=final.append(df)
```
